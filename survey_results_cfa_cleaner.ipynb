```python
import pandas as pd
```

```python
values_df = pd.read_csv('data/fall-25/sos-values_fall-2025.csv')
experiences_df = pd.read_csv('data/fall-25/experiences-survey-results-fall-25.csv')
```

```python
VALUES_SOS_COLUMNS = [
  'Q1_1',
  'Q1_2',
  'Q1_3',
  'Q1_4',
  'Q5_1',
  'Q5_2',
  'Q5_3',
  'Q5_4',
  'Q5_5',
  'Q10_1',
  'Q10_2',
  'Q10_3',
  'Q10_4',
  'Q10_5',
  'Q10_6',
]

EXPERIENCES_SOS_COLUMNS = [
  'q_climate_1',
  'q_climate_2',
  'q_climate_3',
  'q_climate_4',
  'q_structure_1',
  'q_structure_2',
  'q_structure_3',
  'q_structure_4',
  'q_structure_5',
  'q_structure_6',
  'q_vibrancy_1',
  'q_vibrancy_2',
  'q_vibrancy_3',
  'q_vibrancy_4',
  'q_vibrancy_5',
  'q_vibrancy_6',
]
```

```python
def clean_values_experiences_df(df, columns_to_select):
  df_valid_response = df[df['DistributionChannel'] == 'anonymous'].copy()
  df_clean = df_valid_response[2:]
  df_sos_responses = df_clean[columns_to_select]

  rows_to_drop = []
  # Drop rows that have only NaN for SOS responses (no response) and less than 80%
  # of items filled out
  nan_df = df_sos_responses.isnull()
  all_nan_df = nan_df.sum(axis=1)
  print(all_nan_df)
  all_nan_rows = all_nan_df[all_nan_df == nan_df.shape[1]].index
  less_than_eighty_filled_rows = all_nan_df[all_nan_df >= nan_df.shape[1] * 0.2].index
  rows_to_drop += list(all_nan_rows)
  rows_to_drop += list(less_than_eighty_filled_rows)

  # Drop rows with response time < 60 seconds
  df_clean['Duration (in seconds)'] = pd.to_numeric(df_clean['Duration (in seconds)'])
  low_response_drop_rows = df_clean[df_clean['Duration (in seconds)'] < 60].index
  rows_to_drop += list(low_response_drop_rows)

  return df_sos_responses.drop(rows_to_drop)

def write_cleaned_df(out_dir, out_name, df):
  df.to_csv(f'{out_dir}/{out_name}', index=False)
```

```python
cleaned_values_df = clean_values_experiences_df(values_df, VALUES_SOS_COLUMNS)
cleaned_experiences_df = clean_values_experiences_df(experiences_df, EXPERIENCES_SOS_COLUMNS)
write_cleaned_df('data/fall-25', 'values-cfa-cleaned.csv', cleaned_values_df)
write_cleaned_df('data/fall-25', 'experiences-cfa-cleaned.csv', cleaned_experiences_df)
```
